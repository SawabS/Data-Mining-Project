### 2.3. Candidate datetime and numeric columns detection (heuristics)

```python
def detect_columns(df):
    """Automatically detect important column types with validation"""
    
    # Heuristic detection
    candidate_dt = [c for c in df.columns if any(keyword in c.lower() for keyword in 
                      ["time", "date", "timestamp"])]
    candidate_lat = [c for c in df.columns if "lat" in c.lower()]
    candidate_lng = [c for c in df.columns if any(keyword in c.lower() for keyword in 
                      ["lng", "long", "lon"])]
    candidate_num = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
    
    # Common field validation
    common_fields = [
        "ID","Severity","Start_Time","End_Time","Start_Lat","Start_Lng",
        "Distance(mi)","City","State","Weather_Condition","Temperature(F)"
    ]
    found_common = [cf for cf in common_fields if cf in df.columns]
    missing_common = [cf for cf in common_fields if cf not in df.columns]
    
    # Results with confidence scoring
    results = {
        'datetime_cols': candidate_dt,
        'latitude_cols': candidate_lat, 
        'longitude_cols': candidate_lng,
        'numeric_cols': candidate_num,
        'found_common_fields': found_common,
        'missing_common_fields': missing_common,
        'confidence': len(found_common) / len(common_fields) if common_fields else 0
    }
    
    return results

# Usage
column_analysis = detect_columns(df)
print(f"Dataset confidence: {column_analysis['confidence']:.1%}")
print(f"Missing key fields: {column_analysis['missing_common_fields']}")
```

### 2.4. Convert datetimes safely

```python
def safe_to_datetime(s, utc=False):
    try:
        return pd.to_datetime(s, errors='coerce', utc=utc, infer_datetime_format=True)
    except Exception:
        return pd.to_datetime(s, errors='coerce', utc=utc)

candidate_dt = column_analysis.get('datetime_cols', [])

# Excluding "Timezone" from candidate datetime columns
candidate_dt_filtered = [c for c in candidate_dt if c != "Timezone"]

# Applying safe conversion
for c in candidate_dt_filtered:
    df[c] = safe_to_datetime(df[c])

# Persist inferred datetime conversion results
dt_report = df[candidate_dt_filtered].dtypes
dt_report.to_frame(name='dtype').to_csv(os.path.join(SAVE_DIR, "02_datetime_inference.csv"))
dt_report
```

### 2.5. Numerics' downcast to reduce memory

```python
def downcast_numeric(series):
    if pd.api.types.is_float_dtype(series):
        return pd.to_numeric(series, downcast='float')
    if pd.api.types.is_integer_dtype(series):
        return pd.to_numeric(series, downcast='integer')
    return series

for c in df.columns:
    df[c] = downcast_numeric(df[c])

mem_total_opt = df.memory_usage(deep=True).sum()
print(f"Memory after downcast: {mem_total_opt/1e6:.2f} MB ({(mem_total_opt/mem_total-1)*100:.1f}% change)")
```

### 3.4. Plausible ranges and sanity checks

```python
checks = {}

if 'Severity' in df.columns:
    checks['Severity_range'] = df['Severity'].describe()
    # Common severity range: 1-4
    invalid_severity = df['Severity'][~df['Severity'].isin([1,2,3,4])]
    checks['Severity_out_of_set'] = invalid_severity.shape[0]

if 'Distance(mi)' in df.columns:
    dist = df['Distance(mi)']
    checks['Distance_stats'] = dist.describe()
    checks['Distance_negative'] = (dist < 0).sum()
    checks['Distance_extreme'] = (dist > 1000).sum()

# Lat/Lng plausible bounds
for lat_col in [c for c in df.columns if 'lat' in c.lower()]:
    checks[f'{lat_col}_out_of_bounds'] = (~df[lat_col].between(-90, 90)).sum()
for lng_col in [c for c in df.columns if 'lng' in c.lower() or 'long' in c.lower()]:
    checks[f'{lng_col}_out_of_bounds'] = (~df[lng_col].between(-180, 180)).sum()

# Weather plausibility checks
for wc in ['Temperature(F)','Wind_Speed(mph)','Visibility(mi)','Humidity(%)','Pressure(in)']:
    if wc in df.columns:
        s = df[wc]
        checks[f'{wc}_negatives'] = (s < 0).sum() if pd.api.types.is_numeric_dtype(s) else None  # Removed int() wrapper

pd.Series(checks).to_csv(os.path.join(SAVE_DIR, "04_range_checks.csv"))
pd.Series(checks)
```

## 4. Temporal and spatial validation

### 4.1. Year coverage and monthly distribution

```python
def year_month_counts(df, time_col):
    s = df[time_col].dropna()
    return (s.dt.to_period('M')
              .value_counts()
              .sort_index()
              .rename_axis('year_month')
              .reset_index(name='count'))

time_col = 'Start_Time' if 'Start_Time' in df.columns else (candidate_dt[0] if candidate_dt else None)

if time_col:
    ym = year_month_counts(df, time_col)
    # Period to string and replacing the original column
    ym['year_month'] = ym['year_month'].astype(str)
    
    ym.to_csv(os.path.join(SAVE_DIR, "05_year_month_counts.csv"), index=False)
    display(ym.head(12))
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=ym, x='year_month', y='count')
    plt.xticks(rotation=45, ha='right')
    plt.title(f"Monthly accident counts ({time_col})")
    
    # x-axis spacing by showing fewer labels
    n = len(ym)
    if n > 24:
        step = 3
    elif n > 12:
        step = 2
    else:
        step = 1
    
    plt.xticks(range(0, n, step), ym['year_month'].iloc[::step])
    
    plt.tight_layout()
    plt.show()
else:
    print("No datetime column found for temporal analysis.")
```

### 4.2. Duration sanity (Start_Time vs End_Time)

```python
if all(c in df.columns for c in ['Start_Time','End_Time']):
    dur = (df['End_Time'] - df['Start_Time']).dt.total_seconds()
    df['Duration_sec'] = dur
    
    # Converting seconds to a readable format for display
    def format_duration(seconds):
        if pd.isna(seconds):
            return "N/A"
        hours = int(seconds // 3600)
        minutes = int((seconds % 3600) // 60)
        secs = int(seconds % 60)
        if hours > 0:
            return f"{hours}h {minutes}m {secs}s"
        elif minutes > 0:
            return f"{minutes}m {secs}s"
        else:
            return f"{secs}s"
    
    # Stats' calculations and formatting them
    duration_stats = dur.describe()
    negative_dur = (dur < 0).sum()
    extreme_dur = (dur > 7*24*3600).sum()  # >7 days heuristic
    
    print("Duration stats:")
    for stat, value in duration_stats.items():
        if stat == 'count':
            print(f"  {stat}: {int(value)}")
        else:
            print(f"  {stat}: {format_duration(value)}")
    
    print(f"Negative durations: {negative_dur} | Extreme durations: {extreme_dur}")
else:
    print("No duration computation possible; missing Start_Time/End_Time.")
```

### 4.4. Categorical profiling

```python
def profile_categorical(series, top_k=20, min_freq=10):
    vc = series.value_counts(dropna=True)
    return {
        'unique': int(series.nunique(dropna=True)),
        'top_values': vc.head(top_k).to_dict(),
        'rare_values': int((vc < min_freq).sum())
    }

cat_cols = [c for c in df.columns if df[c].dtype == 'object' or pd.api.types.is_string_dtype(df[c])]
cat_report = {c: profile_categorical(df[c]) for c in cat_cols}
with open(os.path.join(SAVE_DIR, "06_categorical_profile.json"), 'w') as f:
    json.dump(cat_report, f, indent=2)
list(cat_report.items())[:5]
```

### 4.5. Numeric profiling and outlier flags

```python
num_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
num_profile = {}
outlier_flags = pd.DataFrame(index=df.index)

for c in num_cols:
    s = df[c].dropna()
    
    # Skipping columns that are actually boolean (True/False)
    if s.dtype == bool:
        continue
        
    num_profile[c] = {
        'count': int(s.count()),
        'mean': float(s.mean()),
        'std': float(s.std()),
        'min': float(s.min()) if len(s) else None,
        'q1': float(s.quantile(0.25)) if len(s) else None,
        'median': float(s.median()) if len(s) else None,
        'q3': float(s.quantile(0.75)) if len(s) else None,
        'max': float(s.max()) if len(s) else None,
        'skew': float(stats.skew(s)) if len(s) > 2 else None
    }
    
    # Robust outlier: modified z-score using median absolute deviation
    if len(s) > 0:
        # Ensuring we're working with numeric data
        s_numeric = s.astype(float)
        med = s_numeric.median()
        mad = stats.median_abs_deviation(s_numeric, scale='normal', nan_policy='omit') if s_numeric.std() > 0 else 0
        
        if mad and np.isfinite(mad) and mad > 0:
            mz = 0.6745 * (s_numeric - med) / mad
            # Creating a full-length series with False values, then update with outlier flags
            outlier_series = pd.Series(False, index=df.index)
            outlier_series.loc[s.index] = (mz.abs() > 3.5)
            outlier_flags[f"{c}_outlier"] = outlier_series
        else:
            outlier_flags[f"{c}_outlier"] = False
    else:
        outlier_flags[f"{c}_outlier"] = False

pd.DataFrame(num_profile).T.to_csv(os.path.join(SAVE_DIR, "07_numeric_profile.csv"))
outlier_flags.to_csv(os.path.join(SAVE_DIR, "08_outlier_flags.csv"))
pd.DataFrame(num_profile).T.head(10)
```
